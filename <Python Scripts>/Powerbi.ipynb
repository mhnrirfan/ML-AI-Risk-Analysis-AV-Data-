```python
import pandas as pd

# File paths
US_df_path = '/Users/mahnooriqbal/COMP702 Project/ML-AI-Risk-Analysis-AV-Data-/Datasets/US_imputed_data.csv'
UK_df_path = "/Users/mahnooriqbal/COMP702 Project/ML-AI-Risk-Analysis-AV-Data-/Datasets/UK/UK-cleaned_data.csv"

# Load CSV files into DataFrames
US_df = pd.read_csv(US_df_path)
UK_df = pd.read_csv(UK_df_path)


```

```python
print("US DataFrame:")
US_df.head()
```

```python
print("UK DataFrame:")
UK_df = UK_df.drop(columns=['latitude', 'longitude'])
UK_df.head()
```

```python
common_cols = US_df.columns.intersection(UK_df.columns)

# Select only the common columns in both DataFrames to avoid mismatches
US_common = US_df[common_cols]
UK_common = UK_df[common_cols]

# Concatenate rows
merged_df = pd.concat([US_common, UK_common], axis=0, ignore_index=True)

merged_df.head()
```

```python
merged_df.shape
```

```python
output_file = '/Users/mahnooriqbal/COMP702 Project/ML-AI-Risk-Analysis-AV-Data-/Datasets/combined_risk_analysis_PBi.csv'
merged_df.to_csv(output_file, index=False)
print(f"Merged data saved to {output_file}")
```
